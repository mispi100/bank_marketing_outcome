```python
import pydot
from IPython.display import Image, display
import io
```

```python
graph = pydot.Dot(graph_type='graph')

inputs = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9']
outputs = ['y']

for node_id in inputs:
    graph.add_node(pydot.Node(node_id, shape="circle", style="filled", fillcolor="lightgrey", label=node_id))


for node_id in outputs:
    graph.add_node(pydot.Node(node_id, shape="doublecircle", style="filled", fillcolor="lightgreen", label=node_id))

for inp in inputs:
    for out in outputs:
        graph.add_edge(pydot.Edge(inp, out))

png_str = graph.create_png(prog='dot')
sio = io.BytesIO(png_str)
print("Single-Neuron Model")
display(Image(sio.getvalue()))
```

```python
graph = pydot.Dot(graph_type='graph')

inputs = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9']
hidden = ['h1', 'h2', 'h3', 'h4']
outputs = ['y']

for node_id in inputs:
    graph.add_node(pydot.Node(node_id, shape="circle", style="filled", fillcolor="lightgrey", label=node_id))

for node_id in hidden:
    graph.add_node(pydot.Node(node_id, shape="circle", label=node_id, style="filled", fillcolor="lightgreen"))

for node_id in outputs:
    graph.add_node(pydot.Node(node_id, shape="doublecircle", style="filled", fillcolor="lightgreen", label=node_id))

for inp in inputs:
    for hid in hidden:
        graph.add_edge(pydot.Edge(inp, hid))

for hid in hidden:
    for out in outputs:
        graph.add_edge(pydot.Edge(hid, out))

png_str = graph.create_png(prog='dot')
sio = io.BytesIO(png_str)
print("Single-Hidden-Layer Model")
display(Image(sio.getvalue()))
```
